```python
# Câu 1
import pandas as pd
from tables.table import Column

file = "/content/drive/MyDrive/Khai Phá Dữ Liệu/21 06/Full_Mark_2020.csv"

# Đọc tập tin CSV vào DataFrame
df = pd.read_csv(file)

# Xem thông tin của dataframe
df
```

```python
# Đếm số lượng học sinh dự thi
total_students = len(df)
print("Số lượng học sinh dự thi là: ", total_students)
```

```python
# Câu 2
import numpy as np

# Số bài thi môn Toán
df_math = df[["Code", "Toan"]]

# Xóa thí sinh không thi môn Toán
df_math = df_math.dropna(subset="Toan")

# In số bài thi môn Toán
math_scores = len(df_math)
print("Số bài thi môn Toán là: ", math_scores)

# Lưu vào file data.csv
df_math.to_csv("data.csv")
df_math
```

```python
# Phổ điểm của môn Toán

# Chuyển đổi danh sách thành kiểu số
df_math = pd.to_numeric(df_math['Toan'], errors='coerce')

# Tạo dataframe từ danh sách điểm
df_ThongKe = pd.DataFrame({'Diem' : df_math})

# Thống kê tần suất xuất hiện
frequency_table = df_ThongKe['Diem'].value_counts().sort_index().reset_index()
frequency_table.Column = ['Giá trị điểm', 'Tần suất xuất hiện']

# Hiển thị bảng thống kê
frequency_table
```

```python
import matplotlib.pyplot as plt

# Khoảng điểm
bins = [i / 10 for i in range(0, 101, 2)]

# Vẽ histogram
plt.hist(df_math, bins = bins, edgecolor='black')
plt.xlabel('Điểm')
plt.ylabel('Số lượng thí sinh')
plt.title('Phổ điểm thi môn Toán')
plt.show()

# Điểm trung bình môn Toán
mean_score = df_math.mean().round(3)
print("Điểm trung bình môn Toán:", mean_score)

# Điểm trung vị môn Toán
median_score = df_math.median()
print("Điểm trung vị môn Toán:", median_score)

# Mode (phổ biến nhất) của môn Toán
mode_scores = df_math.mode()
print("Mode của môn Toán:", mode_scores)

# Độ lệch chuẩn của môn Toán
std_deviation = df_math.std().round(3)
print("Độ lệch chuẩn của môn Toán:", std_deviation)
```

```python
# Xây dựng phổ điểm tổng cộng (Toán+Văn+Ngoại ngữ+ KHXH) và tính toán các giá trị trung bình, trung vị, độ lệch chuẩn và mode

import matplotlib.pyplot as plt

# Tính điểm tổng cộng
df_total = df['Toan'] + df['Nguvan'] + df['Ngoaingu'] + df['KHXH']

# Khoảng điểm
bins = range(0, 41)

# Vẽ histogram
plt.hist(df_total, bins = bins, edgecolor = 'black')
plt.xlabel('Điểm tổng cộng (Toán+Văn+Ngoại ngữ+ KHXH)')
plt.ylabel('Số lượng thí sinh')
plt.title('Phổ điểm tổng cộng (Toán+Văn+Ngoại ngữ+ KHXH)')
plt.show()

# Điểm trung bình tổng cộng (Toán+Văn+Ngoại ngữ+ KHXH)
mean_score = df_total.mean().round(3)
print("Điểm trung bình:", mean_score)

# Điểm trung vị tổng cộng (Toán+Văn+Ngoại ngữ+ KHXH)
median_score = df_total.median()
print("Điểm trung vị:", median_score)

# Độ lệch chuẩn tổng cộng (Toán+Văn+Ngoại ngữ+ KHXH)
std_deviation = df_total.std().round(3)
print("Độ lệch chuẩn:", std_deviation)

# Mode (phổ biến nhất) tổng cộng (Toán+Văn+Ngoại ngữ+ KHXH)
mode_scores = df_total.mode()
print("Mode:", mode_scores)
```
